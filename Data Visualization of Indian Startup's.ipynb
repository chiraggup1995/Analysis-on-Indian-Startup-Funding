```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

startup_df = pd.read_csv('C:/Users/chigup01/Downloads/startup_funding.csv')
```

```python
#Now Lets Drive deep into this dataset and look for Amazing Facts About Indian Startups Funding
#Since Remarks seems of no importance we will delete it.
del(startup_df['Remarks'])
```

```python
#Converting the DataType of Column of Data and AmountinUSD
startup_df['AmountInUSD'] = startup_df['AmountInUSD'].apply(lambda x: float(str(x).replace("," , "")))
startup_df['AmountInUSD'] = pd.to_numeric(startup_df['AmountInUSD'])
#startup_df['StartupName'] = startup_df['StartupName'].str
```

```python
startup_df.dtypes
```

```python
#Checking Total number of Missing Value 
startup_df.isnull().sum()
```

```python
top_10 = startup_df.nlargest(10, 'AmountInUSD')
max_funded_amount = top_10['AmountInUSD'].max()
top_funded = top_10[top_10.AmountInUSD == max_funded_amount]
top_startup_name = top_funded.loc[:,'StartupName']
print(top_startup_name)
print("Most Funded Startup of India is with $" ,top_10['AmountInUSD'].max())
```

```python
top_10
```

```python
plt.figure(figsize=(14,6))
plt.bar(top_10['StartupName'], top_10['AmountInUSD'])
plt.title("Top Funded Startup in India")
plt.xlabel("Name of a Company")
plt.ylabel("Dollar Funded")
#plt.yticks(range(1000000000,1600000000,200000000))
plt.show()
```

```python
#Let see least funded startup 
bottom_10 = startup_df.nsmallest(10, 'AmountInUSD')
least_funded_amount = bottom_10['AmountInUSD'].max()
bottom_funded = bottom_10[bottom_10.AmountInUSD == least_funded_amount]
bottom_startup_name = bottom_funded.loc[:,'StartupName']
print(bottom_startup_name)
print("Most Funded Startup of India is with $ " ,bottom_10['AmountInUSD'].max())
```

```python
plt.figure(figsize=(14,6))
plt.bar(bottom_10['StartupName'], bottom_10['AmountInUSD'])
plt.title("Least Funded Startup in India")
plt.xlabel("Name of a Company")
plt.ylabel("Dollar Funded")
#plt.yticks(range(1000000000,1600000000,200000000))
plt.show()
```

```python
#Let See of which sector top funded startup are
plt.figure(figsize=(14,6))
top_scatter = plt.scatter(top_10['IndustryVertical'], top_10['StartupName'])
plt.title("Top Funded Startup in India categorize Sectorwise")
plt.xlabel("Sector Name")
plt.ylabel("Name of Company")
```

```python
#Top Investor Name 

top_10.SubVertical.fillna(top_10.IndustryVertical, inplace=True)
plt.figure(figsize=(10,6))
subvertical = list(top_10['SubVertical'])
investor_name = list(top_10['InvestorsName'])
plt.scatter(subvertical, investor_name)

#Annotating the Scatter Plot
for i, txt in enumerate(top_10['StartupName']):
    plt.annotate(txt, (subvertical[i], investor_name[i]))
    
plt.title("Top Sectors that are funded by Investors")
plt.xlabel("Sector Name")
plt.ylabel("Investor Name")
plt.xticks(rotation='vertical')
plt.show()
```

```python
#Lets Check which city is playing major role in building startup

cites_count = startup_df['CityLocation'].value_counts()
top_15_cities = cites_count.nlargest(15)

#Total Number of startup funded
total_number_startup = len(startup_df['StartupName'].unique())

#Calculating Percentage wise contribution of each City
city_perc = []
for i in (range(len(top_15_cities))):
    percentage = ((top_15_cities.values[i] / total_number_startup) * 100)
    percentage = round(percentage,2)
    city_perc.append(percentage)

#city_perc = city_perc[::-1]

plt.figure(figsize=(10,6))
ax = top_15_cities.plot(kind='barh', figsize=(10,7),
                                        color="coral", fontsize=13);

# Draw vertical axis lines
        
plt.xlabel('Total Number of Startup', labelpad=20, weight='bold', size=12)
plt.ylabel('Location', labelpad=20, weight='bold', size=12)
count = 0
for i in ax.patches:
    ax.text(i.get_width()+.1, i.get_y()+.10, city_perc[count], fontsize=13, color='dimgrey')
    count= count + 1
for i in range(0,700,100):
    plt.axvline(i, color='grey', alpha=0.25)
plt.title('Major Cities where Startup Industry is building')
plt.show()
```

```python
#Let check which year was prominent in getting high number of Funding
startup_df['Date'][startup_df['Date']=='12/05.2015'] = '12/05/2015'
startup_df['Date'][startup_df['Date']=='13/04.2015'] = '13/04/2015'
startup_df['Date'][startup_df['Date']=='15/01.2015'] = '15/01/2015'
startup_df['Date'][startup_df['Date']=='22/01//2015'] = '22/01/2015' 
startup_df['year'] = (pd.to_datetime(startup_df['Date'],format='%d/%m/%Y', errors ='ignore').dt.year)
#+(pd.to_datetime(startup_df['Date'],format='%d/%m/%Y').dt.month)
```

```python
num_startup_yearwise = startup_df['year'].value_counts()
num_startup_yearwise.plot.pie(y='Year', figsize=(5, 5), autopct='%1.1f%%')
```

```python
#Year_Month Distribution of Number of Startup Funded
startup_df['year_month'] = (pd.to_datetime(startup_df['Date'],format='%d/%m/%Y', errors ='ignore').dt.year*100) + (pd.to_datetime(startup_df['Date'],format='%d/%m/%Y').dt.month)
year_month = (startup_df['year_month'].value_counts())
plt.figure(figsize=(12,10))
sns.barplot(year_month.index, year_month.values, alpha=0.9, color = "coral")
plt.xticks(rotation='vertical')
plt.xlabel('Year-Month', fontsize=12)
plt.ylabel('Number of fundings made', fontsize=12)
plt.title("Year-Month Distribution of number of Sartup Fnded", size= 18)
plt.show()
```
